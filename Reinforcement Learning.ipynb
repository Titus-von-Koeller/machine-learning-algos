```python
from __future__ import division

legend = {'#': 0, 
          '.': 1, 
          'S': 1, 
          'G': 1}

grid_input = '''
    .....#G
    .....#.
    .##....
    .S#....
    .......
    '''

gamma = 0.9  # discount factor
alpha =  .4  # learning-rate (for Q-learning)
```

```python
from numpy import array

# top left is (0,0)
# transpose so tile @ (x,y) == grid[x,y]
grid_s = array([[tile for tile in row.strip()] 
                      for row in grid_input.strip().split()]).T

reward = array([[10 if tile == 'G' else 0
                             for tile in row.strip()] 
                             for row in grid_input.strip().split()]).T

grid = array([[legend[tile] for tile in row.strip()] 
                            for row in grid_input.strip().split()]).T

positions = states = set((x,y) for y,row in enumerate(grid_input.strip().split())
                               for x,tile in enumerate(row.strip())
                               if grid[x,y] > 0)

start = next((x,y) for y,row in enumerate(grid_input.strip().split())
                   for x,tile in enumerate(row.strip())
                   if tile == 'S')

goal = next((x,y) for y,row in enumerate(grid_input.strip().split())
                  for x,tile in enumerate(row.strip())
                  if tile == 'G')

assert grid_s[start] == 'S'
assert grid_s[goal] == 'G'
```

```python
actions = set('NSEW')

dirs = {'N': (+0, -1), 
        'S': (+0, +1), 
        'E': (+1, +0),      
        'W': (-1, +0),}

def move(grid, pos, dir):
    new_pos = x, y = tuple(a+b for a,b in zip(pos, dirs[dir]))
  
    # can't travel outside of grid
    if new_pos not in states:
        return pos
    
    # can't travel to blacked-out space
    if grid[new_pos] == 0:
        return pos
    
    return new_pos
    
def allowed(grid, pos, dir):
    new_pos = tuple(a+b for a,b in zip(pos, dirs[dir]))
    return new_pos in states and grid[new_pos] != 0

def allowed_actions(grid, pos):
    rv = [action for action in actions if allowed(grid, pos, action)]
    shuffle(rv)
    return rv    
    
assert move(grid, start, 'N') == start
assert move(grid, goal, 'N') == goal
assert move(grid, goal, 'W') == goal
assert move(grid, move(grid, goal, 'S'), 'N') == goal

pos = start

#for d in 'WNNEEEESEENN':
#    pos = move(grid, pos, d)
#assert pos == goal
```

```python
def pairwise(g):
    g = iter(g)
    x, y = next(g), next(g)
    while True:
        yield x, y
        x, y = y, next(g)

def dict_differ(d1, d2, threshold=0.0001):
    return set(d1).symmetric_difference(d2) or \
           any(abs(d1[k] - d2[k]) > threshold for k in (set(d1)|set(d2)))
```

# Policy Evaluation

```python
from IPython.display import display
from random import choice, shuffle
from numpy import argmax
from itertools import islice

def transition(grid, state, action):    
    yield 0.7, move(grid, state, action)
    for a in actions - {action,}:
        yield 0.1, move(grid, state, a)

def expected_utility(V, grid, state, action):
    return sum(p * V[s1] for p, s1 in transition(grid, state, action))
        
def policy_evaluation(V, pi, grid, gamma, k=20):
    for _ in range(k):
        for s in states:
            V[s] = reward[s] + gamma * expected_utility(V, grid, s, pi[s])
    return V
      
def policy_iteration(grid, gamma):
    def iterate(V, pi):
        while True:
            V = policy_evaluation(V, pi, grid, gamma)
            for s in states:
                pi[s] = max(actions, key=lambda a: expected_utility(V, grid, s, a))                
            yield V
    
    iteration = iterate(V = {s: 0 for s in states}, 
                        pi = {s: choice(list(actions)) for s in states})    
    
    return next((step, V1) for step, (V1, V2) in islice(enumerate(pairwise(iteration)),100,None)
                                           if not dict_differ(V1, V2))
```

# Q-Learning

```python
from random import choice, shuffle
from itertools import count
from numpy.random import random
   
def qlearning(Q, grid, alpha, gamma, steps=100):    
    accumulated_reward = 0
    pos = start
    
    for _ in range(steps):
        if pos == goal:
            break
            
        # ε-greedy policy:
        #   randomly choose action, 97-100% chance of best action, 0-3% chance of any other action
        best_action = max(allowed_actions(grid, pos), key=lambda action: Q[pos, action])        
        action = choice([best_action]*(99-len(allowed_actions(grid, pos))) + allowed_actions(grid, pos))

        # take action a, observe reward, new state
        new_pos = move(grid, pos, action)        
        r = reward[new_pos] # reward for action
        
        accumulated_reward += r
        
        # calculates the best Q-value given the best action from the new position
        best_new_action = max(allowed_actions(grid, new_pos), key=lambda action: Q[new_pos, action])         
        
        Q[pos, action] += alpha*(r + gamma * Q[new_pos, best_new_action] - Q[pos, action])
        pos = new_pos
    
    return accumulated_reward
```

## Question 2a.

```python
Q = {(pos, action): 0 for pos in positions
                      for action in allowed_actions(grid, pos)}
rewards = []

for num_steps in count(1):
    Q_before = Q.copy()
    accumulated_reward = qlearning(Q, grid, alpha, gamma)
    Q_after = Q.copy()
    rewards.append(rewards[-1] + accumulated_reward if rewards else accumulated_reward)
    
    if num_steps > 50 and all(abs(Q_before[k] - Q_after[k]) < 0.0001 for k in Q):
        break
    
display("Converged in {} steps.".format(num_steps))
```
